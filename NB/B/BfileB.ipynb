````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
````

Create a normal distribution with mean as $\mu$ and standard deviation as $\sigma.$

````python
mu = 1500
std = 300
snd = stats.norm(mu, std)
````

Generate 1000 random values between $a$, $b.$ We will know these limites after plotting.

````python
x = np.linspace(-500, 5000, 1000)
````

Plot the standard normal distribution for different values of random variable
falling in the range $a, b$

````python
plt.figure(figsize=(12, 6));
plt.figure(figsize=(7.5,7.5))
plt.plot(x, snd.pdf(x))
plt.xlim(200, 3000)
plt.title(r'Normal Distribution $(\mu= 1500, \sigma = 300)$', fontsize='15')
plt.xlabel('Values of Random Variable X', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
````

The ```cdf``` method calculates the proportion of a normally distributed population that is less than or equal to a given value. (CDF stands for Cumulative Distribution Function).

```cdf``` function calculates the probability for a given normal distribution value,

````python
snd.cdf(1800)
````

```ppf()``` function calculates the normal distribution value for which a given probability is the required value. 

````python
snd.ppf(0.85)
````

````python
plt.figure(figsize=(12, 6))
shaded_region =  (x <= 1800)
plt.fill_between(x, snd.pdf(x), alpha=0.2)
plt.shaded_region =  (x <= 1800)
plt.fill_between(x, snd.pdf(x), where=shaded_region)
plt.xlim(200, 3000)
plt.title(r'Normal Distribution $(\mu= 1500, \sigma = 300)$', fontsize='15')
plt.xlabel('Values of Random Variable X', fontsize='15')
plt.ylabel('CDF', fontsize='15');
````
